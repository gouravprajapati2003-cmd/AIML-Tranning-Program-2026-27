```python
def my_mean_squared_error (x_pred, y_pred) :
    sum = 0;
    for i in range(len(x_pred)):
        sum = sum + (x_pred[i] - y_pred[i]) * (x_pred[i] - y_pred[i])
        result = sum / len(y_pred)
        return result
```

```python
mse1 = my_mean_squared_error(y_test_arr, y_pred)
```
