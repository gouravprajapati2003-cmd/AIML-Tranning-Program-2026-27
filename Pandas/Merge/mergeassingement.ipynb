Write a Pandas program to merge two DataFrames on a single column.

DataFrames on a single column. 

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})
df1
```

```python
df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
})
df2
```

```python
merged_data = pd.merge(df1, df2,  on ='ID')
merged_data
```
