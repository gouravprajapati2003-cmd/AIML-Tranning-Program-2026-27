```python
df = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth ': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'age': [12, 12, 13, 13, 14, 12],
    'height': [173, 192, 186, 167, 151, 159],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']
})
df                  
```

## WAP to find out the size of school code

```python
result = df.groupby('school_code').size()
result
```

## WAP to find out the Maximum age from School Wise

```python
result = df.groupby('school_code')['age'].max()
```

```python
result = df.groupby('school_code').agg({'age' : ['max']})
```

## Write a pandas program to spilt the following dataframe by school spde and get mean, min and max value of age for each school

```python
result = df.groupby('school_code').agg({'age' : ['mean', 'min', 'max']})
result
```

## Write a pandas program to spilt the following dataframe by school code, class code and get mean, min and max value of age for each school

```python
result = df.groupby(['school_code','class']).agg({'age' : ['mean', 'min', 'max']})
result
```

```python
result = df.groupby(['school_code', 'class']).agg({'age' : ['mean', 'min', 'max'], 'height' : ['mean', 'min', 'max']})
result
```
