```python
def updateSize(SIZE):
    temp = SIZE.split(" ")
    if( int(temp[0]) > 5):
        return 3
    else:
        return temp[0]
```

```python
df['newsize'] = df['SIZE'].apply(updateSize)
df
```

```python
def updatePrice(price):
    temp = price.split("-")
    if(len(temp) == 2):
        return (float(temp[0]) + float(temp[1]))/2
    else:
        return float(temp[0])
```

```python
df['price'] = df['price'].apply(updatePrice)
df
```
