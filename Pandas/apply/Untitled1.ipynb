```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Bengaluru_House_Data.csv")
df
```

## Print no of rows and columns

```python
rows, columns = df.shape
print("Rows =", rows)
print("Columns =", columns)
```

## Print the all columns using the loop

```python
for col in df.columns:
    print(f"Column: {col}")
    print(df[col])
    print()
```

## Explore all the columns using value_counts() method df['column'].value_counts()

```python
for col in df.columns:
    print("\nColumn:", col)
    print(df[col].value_counts())
```

## Drop these columns area_type', 'availabilty', 'balcony', 'society' using df.drop(columns = [])

```python
df.drop(columns=['area_type', 'availability', 'balcony', 'society'], inplace=True)
```
