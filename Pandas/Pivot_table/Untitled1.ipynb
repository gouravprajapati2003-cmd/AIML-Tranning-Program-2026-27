## Write a Pandas Program to create a pivot table and find the total sale amount region wise, manager wise, salesman wise where Manager = "Doughals"

```python
import pandas as pd
import numpy as np

# Convert Sale_amt to numeric (if it contains commas)
df["Sale_amt"] = (
    df["Sale_amt"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .astype(float)
)

# Filter data for Manager = "Doughals"
df1 = df[df["Manager"] == "Doughals"]

# Create Pivot Table
table = pd.pivot_table(
    df1,
    index=["Region", "Manager", "SalesMan"],
    values="Sale_amt",
    aggfunc=np.sum,
    fill_value=0,
    margins=True
)

print(table)
```

```python
table = pd.pivot_table(
    df[df["Manager"] == "Doughals"],
    index=["Region", "Manager", "SalesMan"],
    values="Sale_amt",
    aggfunc="sum",
)

table
```

```python
table = pd.pivot_table(
    df[df["Manager"] == "Douglas"],
    index=["Region", "Manager", "SalesMan"],
    values="Sale_amt",
    aggfunc="sum",
    fill_value=0,
    margins=True
)

print(table)
```

## Write a Pandas Program to create a pivot table and find the region wise television and home theater sold

```python
df1 = df[df["Item"].isin(["Television", "Home Theater"])]
```

```python
table = pd.pivot_table(
    df1,
    index="Region",
    columns="Item",
    values="Units",
    aggfunc="sum",
    fill_value=0,
    margins=True
)

print(table)
```
