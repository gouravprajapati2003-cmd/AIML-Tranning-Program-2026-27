```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'Product': ['Carrots', 'Broccoli', 'Banana', 'Banana',
                               'Beans', 'Orange', 'Broccoli', 'Banana'],
                   'Category': ['Vegetable', 'Vegetable', 'Fruit', 'Fruit',
                                'Vegetable', 'Fruit', 'Vegetable', 'Fruit'],
                   'Quantity': [8, 5, 3, 4, 5, 9, 11, 8],
                   'Amount': [270, 239, 617, 384, 626, 610, 62, 90]})
df
```

## Get the Total Sales of Each product

```python
result = pd.pivot_table(df, index = ['Product'], values = ['Amount'], aggfunc = [np.sum])
```

```python
result
```

## Get the Total Sales of  Each Product

```python
result = pd.pivot_table(df, index = ['Category'], values = ['Amount'], aggfunc = [np.sum])
```

```python
result
```

## Get the Total Sales by category and Product both

```python
result = pd.pivot_table(df, index = ['Product','Category'], values = ['Amount'], aggfunc = [np.sum])
```

```python
result
```

## Get the Mean, Median, Minimum Sale by Category

```python
result = pd.pivot_table(df, index = ['Category'], values = ['Amount'], aggfunc = [np.mean, np.median, np.min])
```

```python
result
```
